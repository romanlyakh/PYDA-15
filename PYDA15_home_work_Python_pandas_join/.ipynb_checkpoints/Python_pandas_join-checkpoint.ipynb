**Задание 1**

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')

average_lifetime = ratings.groupby('userId').agg(['max', 'min', 'count'])
average_lifetime = average_lifetime.loc[average_lifetime['timestamp']['count'] > 100]
average_lifetime['time_life'] = average_lifetime['timestamp']['max'] - average_lifetime['timestamp']['min']
average_lifetime.loc[:, ['userId', 'time_life']]
```
