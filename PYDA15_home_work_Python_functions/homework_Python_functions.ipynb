```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2', '5455 028765'],
    '2': ['10006', '5400 028765', '5455 002299'],
    '3': []
}


def owner_document():
    num = input('Введите номер документа:')
    for item in documents:
        if num == item['number']:
            print ('Владелец документа:', item['name'])
            return
    print('Документ не найден в базе')
    
#owner_document()


def find_shelf(shelf):
    num_doc = input('введите номер документа: ')
    for key, value in shelf.items():
        for i in value:
            if i == num_doc:
                print('Документ хранится на полке:', key)
                return
    print('Документ не найден в базе')

#find_shelf(directories)

def info_all_doc(docs, direct):
    for i in(docs):
        for key, value in direct.items():
            if i['number'] in value:
                print(f'№:{i["number"]}, тип: {i["type"]}, владелец: {i["name"]}, полка хранения: {key}')
                

#info_all_doc(documents,directories)

def add_shelf(shelf):
    num_new_shelf = input('Введите номер полки: ')
    if num_new_shelf not in shelf.keys():
        shelf.setdefault(num_new_shelf, [])
        print('Полка добавлена. Текущий перечень полок:',', '.join(shelf.keys()))
    else: print('Такая полка уже существует. Текущий перечень полок:',', ' .join(shelf.keys()))
   

#add_shelf(directories)

def del_shelf(shelf):
    num_delete_shelf = input('Введите номер полки: ')
    if num_delete_shelf not in shelf:
        print('Такой полки не существует. Текущий перечень полок:',', '.join(shelf.keys()))
    elif num_delete_shelf in shelf and not shelf[num_delete_shelf]:
        del(shelf[num_delete_shelf])
        print('Полка удалена. Текущий перечень полок:',', '.join(shelf.keys()))
    else: print('На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: ',', '.join(shelf.keys()))   

#del_shelf(directories)

def main_input():
    while True:
        user_command = input('Введите команду: ')
        if user_command == 'p':
            owner_document()
        elif user_command == 's':
            find_shelf(directories)
        elif user_command == 'l':
            info_all_doc(documents)
        elif user_command == 'as':
            add_shelf(directories)
        elif user_command == 'ds':
            del_shelf(directories)
        elif user_command == 'q':
             print('Выход из программы')
             break
                
main_input()
        
```
